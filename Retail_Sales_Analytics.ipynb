# STEP 5 – Feature Engineering

```python
df.columns
```

```python
# Convert date into proper date format

df['Order.Date'] = pd.to_datetime(df['Order.Date'])
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])
```

```python
df.dtypes
```

```python
# Create Month column
# .dt is used to get date details like month, year, day and quarter

df['Month'] = df['Order.Date'].dt.month_name()
```

```python
# Create year column
df['Year'] = df['Order.Date'].dt.year
```

```python
# Create Day Name column
# Used to analyze which day receives the highest number of orders

df['Day'] = df['Order.Date'].dt.day_name()
```

```python
# Get quarter from date
# Q1=Jan-Mar, Q2=Apr-Jun, Q3=Jul-Sep, Q4=Oct-Dec

df['Quarter'] = df['Order.Date'].dt.quarter
```

```python
# Calculate delivery days
# Find days between order and shipping

df['Shipping Days'] = (df['Ship.Date'] - df['Order.Date']).dt.days
```

```python
df.columns
```

#  Step 6 : Exploratory Data Analysis (EDA)

```python
# Total sales
total_sales = df['Sales'].sum()
print("Total Sales :", total_sales)

# Total profit
total_profit = df['Profit'].sum()
print("Total Profit :", total_profit)

# Total orders
total_orders = df['Order.ID'].nunique()
print("Total Orders :", total_orders)

# Total customers
total_customers = df['Customer.ID'].nunique()
print("Total Customers :", total_customers)

# Calculate average sales
average_sales = df['Sales'].mean()
print("Average Sales :", round(average_sales,2))

# Find highest sale
highest_sale = df['Sales'].max()
print("Highest Sale :", highest_sale)

# Find lowest sale
lowest_sale = df['Sales'].min()
print("Lowest Sale :", lowest_sale)
```

# Step 7 : Data Visualization

```python
# Group sales by month

monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales

# Explanation:
# groupby() groups all same months together.
# sum() adds the sales of each month.
# This shows total sales for every month.

# Business Use:
# Helps identify which month has the highest and lowest sales.
```

```python
# Monthly sales chart

monthly_sales_df = monthly_sales.reset_index()

fig = px.line(
    monthly_sales_df,
    x='Month',
    y='Sales',
    title='Monthly Sales Trend'
)

fig.show()
```

```python
# Group sales by category

category_sales = df.groupby('Category')['Sales'].sum()
category_sales

# Explanation:
# Groups all categories together.
# Calculates total sales of each category.

# Business Use:
# Helps identify the best-selling category.
```

```python
# Category chart

fig = px.bar(
    x=category_sales.index,
    y=category_sales.values,
    title="Sales by Category"
)

fig.show()

# Explanation:
# x shows category names.
# y shows total sales.
# This chart compares sales of each category.

# Business Use:
# Helps find the best-selling category.
```

```python
# Group sales by product

top_products = df.groupby('Product.Name')['Sales'].sum()

# Sort products by sales

top_products = top_products.sort_values(ascending=False)

# Show top 10 products

top_products = top_products.head(10)

top_products

# Explanation:
# Groups sales for each product.
# Sorts products from highest to lowest sales.
# Displays the top 10 products.

# Business Use:
# Helps identify the company's best-selling products.
```

```python
# Product chart

fig = px.bar(
    x=top_products.index,
    y=top_products.values,
    title="Top 10 Products by Sales"
)

fig.show()

# Explanation:
# x = Product names
# y = Total sales
# Bar chart compares the sales of the top 10 products.

# Business Use:
# Helps identify the best-selling products.
# Company can focus on these products to increase revenue.
```
